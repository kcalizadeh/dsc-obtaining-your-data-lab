````python
# Your code here
import sqlite3
import custom_functions as cf

conn = sqlite3.connect('lego.db')
cur = conn.cursor()

query = '''select name from sqlite_master where type='table' order by name;'''
print(list(cur.execute(query)))

cur.execute('''select * from product_details''')

deets = cf.make_frame_from_query(cur, '''select * from product_details''')
info = cf.make_frame_from_query(cur, '''select * from product_info''')
price = cf.make_frame_from_query(cur, '''select * from product_pricing''')
reviews = cf.make_frame_from_query(cur, '''select * from product_reviews''')
deets.head(), info.head(), price.head(), reviews.head()

df = deets.merge(info, on='prod_id').merge(price, on='prod_id').merge(reviews, on='prod_id')
df.head()
````
